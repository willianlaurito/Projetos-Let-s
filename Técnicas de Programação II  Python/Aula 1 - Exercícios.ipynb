## Exercício 2

**a)** Utilizando a base de dados de covid, crie um gráfico de linhas mostrando a evolução do número de casos novos mês a mês, para as DRS 'Grande São Paulo', 'São José do Rio Preto' e 'Marília'. Identifique os DRS por cores e legenda. Não esqueça de colocar título para o gráfico e para os eixos. 

```python
df2.head()
```

```python
df2_p = df2.pivot_table(index="ano_mes", columns="nome_drs", values="casos_novos", aggfunc="sum") 
dfp = df2_p[["São José do Rio Preto", "Marília", "Grande São Paulo"]]
```

```python
dfp.head()
```

```python
grafo = dfp.plot(title="Covid", xlabel="Ano / mês", ylabel="Novos casos de Covid", figsize=(15,10))
grafo.legend(title="Departamento Regional de Saúde");
```

**b)** Repita o gráfico do item a trocando as cores dos DRS.

```python
grafo = dfp.plot(title="Covid", xlabel="Ano / mês", ylabel="Novos casos de Covid", figsize=(15,10),
                color={"São José do Rio Preto" : "#0000FF", "Marília" : "#008000", "Grande São Paulo" : "#8B008B"}, 
                 linestyle='dashdot')
grafo.legend(title="Departamento Regional de Saúde");
```

**c)** Considerando apenas os DRS 'Bauru', 'São José do Rio Preto' e 'Marília', faça três gráficos de pizza, para os meses de abril, maio e junho, mostrando a porcentagem de novos óbitos em cada DRS.

```python
drs = df2.pivot_table(index="ano_mes", columns= "nome_drs", values="obitos_novos", aggfunc="sum")
drs = drs[2:][["Marília", "Bauru", "São José do Rio Preto"]]
```

```python
ax = drs.plot(subplots=True, kind="pie", figsize=(20,15), title="óbitos DO Covid-19",
              autopct="%1.0f%%", labeldistance=None,
              colors={"#00BFFF", "#32CD32", "#FF0000"}, fontsize=20)
```
